## Packages and Function

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Color = '#965da7'
NASAColor = '#84a955'
NASA2Color = '#bc5d41'

sns.set()

#plt.style.use('fivethirtyeight')
def NASAPlotFunction(NasaXList, NasaYList, XList, YList, Constant):
    XListSeconds = np.array(XList)/30
    XListOffset = XListSeconds + Constant
    plt.figure()
    plt.plot(NasaXList,NasaYList, label='NASA', marker='o', color=NASAColor,linewidth=3)
    plt.plot(XListOffset,YList, label='Russell',marker='o', color=Color,linewidth=3)
    #plt.legend()
    #plt.grid(False)
    #plt.yticks(np.arange(int(min(NasaYList)), int(max(NasaYList))+1, .5))
    #plt.tick_params(axis = 'both', which = 'major')
    #plt.box(on=None)
    #plt.grid(color='dimgray')
    plt.xlabel('Time (s)')
    plt.ylabel('Magnitude')
    plt.gca().invert_yaxis()
    return(NasaYList, YList)
    
        
def NASA2PlotFunction(NasaXList, NasaYList, NasaXList2, NasaYList2, XList, YList, Constant1, Constant2):
    XListSeconds = np.array(XList)/30
    XListOffset = XListSeconds + Constant1
    NasaXList2Offset = np.array(NasaXList2) + Constant2
    plt.plot(NasaXList,NasaYList, label='NASA 1', color=NASAColor, linewidth=3)
    plt.plot(NasaXList2Offset,NasaYList2, label='NASA 2', color=NASA2Color,linewidth=3)
    plt.plot(XListOffset,YList, label='Russell',color=Color,linewidth=3)
    #plt.legend()
    #plt.tick_params(axis = 'both', which = 'major')
    #plt.box(on=None)
    #plt.grid(color='dimgray')
    plt.xlabel('Time (s)')
    plt.ylabel('Magnitude')
    plt.gca().invert_yaxis()
    
def PlotFunction(XList, YList, Constant):
    XListSeconds = np.array(XList)/30
    XListOffset = XListSeconds + Constant
    plt.figure()
    plt.plot(XListOffset,YList,linewidth=3,color=Color)
    #plt.legend()
    #plt.grid(False)
    #plt.yticks(np.arange(int(min(NasaYList)), int(max(NasaYList))+1, .5))
    #plt.tick_params(axis = 'both', which = 'major')
    #plt.box(on=None)
    #plt.grid(color='dimgray')
    plt.xlabel('Time (s)')
    plt.ylabel('Magnitude')
    #plt.title('Iridium Flare Light Curve',fontsize=28)
    plt.gca().invert_yaxis()   
    
def RawMagnitudePlotFunction(XList, YList, Constant):
    XListSeconds = np.array(XList)/30
    XListOffset = XListSeconds + Constant
    plt.figure()
    plt.plot(XListOffset,YList,color=Color,linewidth=3)
    #plt.legend()
    #plt.grid(False)
    #plt.yticks(np.arange(int(min(NasaYList)), int(max(NasaYList))+1, .5))
    #plt.tick_params(axis = 'both', which = 'major', labelsize = 14)
    #plt.box(on=None)
    #plt.grid(color='dimgray')
    plt.xlabel('Time (s)')
    plt.ylabel('Pixel Value')

def Residuals(NasaXList,NasaYList,XList,YList,Jump):
    if len(NasaYList) > len(YList):
        CutOff = len(YList)
        NasaYList = (NasaYList[Jump:CutOff+Jump])
        NasaXList = np.array(NasaXList[Jump:CutOff+Jump])
    if len(YList) > len(NasaYList):
        CutOff = len(NasaYList)
        YList = YList[Jump:CutOff+Jump]
        XList = np.array(XList)
        
    NasaYList = np.array(NasaYList)
    YList = np.array(YList)
    Difference = NasaYList - YList
    plt.figure()
    plt.scatter(NasaXList,Difference,color=Color)
    plt.xlabel('Time (s)')
    plt.ylabel('Residual')
    return(NasaXList,Difference)
```
