```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np
from scipy import misc
```

```python
f = misc.imread('m67v.png')
target=(40,146)
subf = f[target[0]-15:target[0]+15,target[1]-15:target[1]+15]
h,w = subf.shape
```

```python
plt.figure(figsize=(15,15))
gs = gridspec.GridSpec(2, 2, width_ratios=[3, 1], height_ratios=[3,1]) 
ax=plt.subplot(gs[0])
plt.imshow(subf, cmap='gray')
sig = plt.Circle((15,15),5,color='cyan', alpha=0.3)
back = plt.Circle((15,15),10, color='red', fill=False)
ax.add_artist(sig)
ax.add_artist(back)

plt.subplot(gs[1])
plt.plot(subf[:,15],np.arange(h,0,-1))
plt.yticks([])

plt.subplot(gs[2])
plt.plot(subf[15,:])
plt.gca().invert_yaxis()
plt.xticks([])

plt.subplot(gs[3])
plt.imshow(f, cmap='gray')
plt.axvline(target[1], color='red')
plt.axhline(target[0], color='red')

plt.tight_layout()
plt.show()
```
